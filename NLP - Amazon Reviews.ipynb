# SENTIMENT ANALYSIS

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
obj = SentimentIntensityAnalyzer()
```

```python
ex1
```

```python
obj.polarity_scores(ex1)
```

```python
from tqdm.notebook import tqdm
```

```python
#Running the Review Polarity Score on the complete Review Dataset
result = {}
for i, row in tqdm(df.iterrows(), total = len(df)):
    textc = row['Text']
    idc = row['Id']
    result[idc] = obj.polarity_scores(textc)
```

```python
result
```

```python
sentiment = pd.DataFrame(result).T
sentiment = sentiment.reset_index().rename(columns={'index': 'Id'})
sentiment = sentiment.merge(df, how = 'left')
```

```python
#Sentiment Analysis merged with Original Data
sentiment
```

## PLOTING SENTIMENT RESULTS

```python
ax = sns.barplot(data = sentiment, x = 'Score', y= 'compound')
ax.set_title('Compound Score based on STAR Reviews')
ax.set_ylim(-0.1, 0.8)
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize = (15, 5))
sns.barplot(data= sentiment, x = 'Score', y = 'pos', ax = axs[0])
sns.barplot(data= sentiment, x = 'Score', y = 'neu', ax = axs[1])
sns.barplot(data= sentiment, x = 'Score', y = 'neg', ax = axs[2])
axs[0].set_title('Positive')
axs[1].set_title('Neutral')
axs[2].set_title('Negative')
plt.show()
```
